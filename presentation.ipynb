## 2. Visualização da Base de Dados Concatenada

```python
path_list = ['NAO_IDENT_null_1988-2020_1988_2005_102142.csv', 'NAO_IDENT_null_1988-2020_2006_2012_102142.csv', 'NAO_IDENT_null_1988-2020_2013_2020_102142.csv']

df_final = pd.DataFrame()

for path in path_list:
    data = pd.read_csv(path, sep=';')
    
    df_final = pd.concat([df_final, data], ignore_index=True)
    
df_final.to_csv('dataSetFinal.csv', index=True)
```

```python
df_final.head()
```

```python
df_final.shape
```

```python
df_final.info()
```

```python
for column in df_final.columns:
    if df_final[column].isnull().mean()*100 > 50:
        print(f"Porcentagem de valores nulos na coluna {column}: {round(df_final[column].isnull().mean()*100, 1)}%")
```

```python
for column in df_final.columns:
    if len(df_final[column].apply(type).unique()) > 1:
        print(f"Coluna {column}: {df_final[column].apply(type).unique()}")
```

## 4. Verificação Temporal

### 4.1 Câncer de Mama

```python
pd.set_option('display.max_rows', None)
```

```python
df_final['Data de Diagnostico'].value_counts()
```

```python
df_final['Data de Diagnostico'].count()
```

```python
df_final['Data de Diagnostico'] = pd.to_datetime(df_final['Data de Diagnostico'], format='%d/%m/%Y')
```

```python
df_final['Data de Diagnostico'].count()
```

```python
df_mama = df_final[df_final['Descrição da Doenca'] == 'MAMA , SOE']

df_mama2 = df_final[df_final['Descrição da Doenca'].isin(['MAMA , SOE', 'QUADRANTE SUPERIOR EXTERNO DA MAMA', 'PROLONGAMENTO AXILAR DA MAMA',
                                                         'QUADRANTE SUPERIOR INTERNO DA MAMA', 'CARCINOMA "IN SITU" DA MAMA, SOE',
                                                         'PORCAO CENTRAL DA MAMA', 'QUADRANTE INFERIOR EXTERNO DA MAMA',
                                                         'LESAO INVASIVA DA MAMA', 'QUADRANTE INFERIOR INTERNO DA MAMA', 'MAMA'])]
```

```python
df_mama2.shape
```

```python
df_mama = df_mama.sort_values(by='Data de Diagnostico')
```

```python
df_mama.shape
```

```python
mama_count = df_mama['Data de Diagnostico'].dt.year.value_counts().sort_index()

plt.plot(mama_count.index, mama_count.values, marker=',', label='Câncer de Mama', linestyle='-')


plt.title("Série Temporal do Câncer de Mama")
plt.ylabel("Quantidade")
plt.xlabel("Data")

plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```
