```python
path = '../data/csv/NAO_IDENT_null_1988-2020_1988_2005_102142.csv'
df = pd.read_csv(path, index_col='Código do Paciente', sep=';')
```

```python
for column in df.columns:
    print(f"Porcentagem de valores nulos na coluna {column}: {round(df[column].isnull().mean()*100, 2)}\n")
```

```python
for column in df.columns:
    if df[column].isnull().mean()*100 > 60 and column != 'Status Vital':
        df.drop(column, axis=1, inplace=True)
```

```python
df['Data de Nascimento'] = pd.to_datetime(df['Data de Nascimento'], format='%d/%m/%Y', errors='coerce')
df['Data do Óbito'] = pd.to_datetime(df['Data de Nascimento'], format='%d/%m/%Y', errors='coerce')
df['Data de Último Contato'] = pd.to_datetime(df['Data de Nascimento'], format='%d/%m/%Y', errors='coerce')
df['Data de Diagnostico'] = pd.to_datetime(df['Data de Nascimento'], format='%d/%m/%Y', errors='coerce')
```

```python
df['Idade'] = pd.to_numeric(df['Idade'], errors='coerce').astype('Int64')
df['Código Profissão'] = pd.to_numeric(df['Código Profissão'], errors='coerce').astype('Int64')
df['Código da Morfologia'] = pd.to_numeric(df['Código da Morfologia'], errors='coerce').astype('Int64')
```

```python
ultimos_4_diagnostico = df['Data de Diagnostico'].str[-4:]
ultimos_4_nascimento = df['Data de Nascimento'].str[-4:]

# Lidar com valores ausentes (NaN)
ultimos_4_diagnostico = ultimos_4_diagnostico.fillna('')
ultimos_4_nascimento = ultimos_4_nascimento.fillna('')

# Calcular a diferença e converter para inteiros
diferenca_idade_calculada = (pd.to_numeric(ultimos_4_diagnostico, errors='coerce') -
                             pd.to_numeric(ultimos_4_nascimento, errors='coerce')).astype('Int64')

# Verificar se a diferença calculada é igual à coluna 'Idade'
df['Verificação'] = (diferenca_idade_calculada == df['Idade']) | ((diferenca_idade_calculada - 1) == df['Idade'])
```
