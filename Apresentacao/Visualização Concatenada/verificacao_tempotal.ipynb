# Verificação Temporal

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
```

```python
path = '../../data/parquet/dataSetFinal.parquet'
df_final = pd.read_parquet(path)
```

```python
df_final['Data de Diagnostico'].count()
```

```python
df_final['Data de Diagnostico'] = pd.to_datetime(df_final['Data de Diagnostico'], format='%d/%m/%Y')
```

## Câncer de Mama

```python
#df_mama = df_final[df_final['Descrição da Doenca'] == 'MAMA , SOE']

df_mama = df_final[df_final['Descrição da Doenca'].isin(['MAMA , SOE', 'QUADRANTE SUPERIOR EXTERNO DA MAMA', 'PROLONGAMENTO AXILAR DA MAMA',
                                                         'QUADRANTE SUPERIOR INTERNO DA MAMA', 'CARCINOMA "IN SITU" DA MAMA, SOE',
                                                         'PORCAO CENTRAL DA MAMA', 'QUADRANTE INFERIOR EXTERNO DA MAMA',
                                                         'LESAO INVASIVA DA MAMA', 'QUADRANTE INFERIOR INTERNO DA MAMA', 'MAMA'])]
```

```python
df_mama = df_mama.sort_values(by='Data de Diagnostico')
```

```python
df_mama.shape
```

```python
df_mama_final = df_mama.set_index('Data de Diagnostico')
```

```python
result_mama = df_mama_final.resample("M")['Código do Paciente'].count()
result_mama.plot()
```
