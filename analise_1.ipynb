```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'NAO_IDENT_null_1988-2020_1988_2005_102142.csv'
df = pd.read_csv(path, index_col='Código do Paciente', sep=';')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
for column in df.columns:
    print(f"Valores únicos em {column}: {df[column].unique()}\n")
```

```python
for column in df.columns:
    print(f"Porcentagem de valores nulos na coluna {column}: {round(df[column].isnull().mean()*100, 2)}\n")
```

```python
for column in df.columns:
    if df[column].isnull().mean()*100 > 60:
        df.drop(column, axis=1, inplace=True)
```

```python
df['Data de Nascimento'] = pd.to_datetime(df['Data de Nascimento'], errors='coerce')
df['Data do Óbito'] = pd.to_datetime(df['Data de Nascimento'], errors='coerce')
df['Data de Último Contato'] = pd.to_datetime(df['Data de Nascimento'], errors='coerce')
df['Data de Diagnostico'] = pd.to_datetime(df['Data de Nascimento'], errors='coerce')
```

```python
df['Idade'] = pd.to_numeric(df['Idade'], errors='coerce').astype('Int64')
df['Código Profissão'] = pd.to_numeric(df['Código Profissão'], errors='coerce').astype('Int64')
df['Código da Morfologia'] = pd.to_numeric(df['Código da Morfologia'], errors='coerce').astype('Int64')
```

```python
print(f"Moda RCBP: {df['Nome do RCBP'].mode().iloc[0]}")
print(f"Moda do sexo dos pacientes: {df['Sexo'].mode().iloc[0]}")
print(f"Moda doenças: {df['Descrição da Doenca'].mode().iloc[0]}")
print(f"Média de idade dos pacientes: {round(df['Idade'].mean(), 2)}")
print(f"Moda dos Códigos de Morfologia: {df['Código da Morfologia'].mode().iloc[0]}")
```

```python
df.info()
```

```python
fig, ax = plt.subplots()

ax.hist('Sexo', data=df)
ax.set_title("Histograma de Sexo dos Pacientes")
ax.set_ylabel("Quantidade de Ocorrências")
ax.set_xlabel("Sexo")

plt.tight_layout()
```

```python
contagem_idades = df['Idade'].value_counts().sort_index()

plt.bar(contagem_idades.index, contagem_idades.values)
plt.xlim(0, 110)

ax.set_title("Scatter de Idade dos Pacientes")
ax.set_ylabel("Quantidade de Ocorrências")
ax.set_xlabel("Idade")

plt.tight_layout()
```

```python
contagem_rcbp = df['Nome do RCBP'].value_counts()

fig, ax = plt.subplots()

ax.bar(contagem_rcbp.head(10).index, contagem_rcbp.head(10).values)
ax.set_title("Histograma de RCBP")
ax.set_ylabel("Quantidade de Ocorrências")
ax.set_xlabel("RCBP")

ax.set_xticks(range(len(contagem_rcbp.head(10).index)))
ax.set_xticklabels(contagem_rcbp.head(10).index, rotation=45, ha='right')

plt.tight_layout()
```
