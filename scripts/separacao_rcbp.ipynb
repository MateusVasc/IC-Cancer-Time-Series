```python
columns_to_drop = ['Código do Paciente', 'Unnamed: 37']

df = df.drop(columns_to_drop, axis=1)
df.columns
```

```python
conditions = [
    df['Nome do RCBP'].isin(['RCBP SALVADOR', 'RCBP TERESINA', 'RCBP RECIFE', 'RCBP ARACAJU', 'RCBP NATAL', 'RCBP JOAO PESSOA', 
                              'RCBP FORTALEZA', 'RCBP ALAGOAS']),
    df['Nome do RCBP'].isin(['RCBP CAMPO GRANDE', 'RCBP MATO GROSSO (INTERIOR)', 'RCBP CUIABA', 'RCBP DISTRITO FEDERAL', 
                              'RCBP GOIÂNIA']),
    df['Nome do RCBP'].isin(['RCBP RORAIMA', 'RCBP BELÉM', 'RCBP PALMAS', 'RCBP MANAUS', 'RCBP ACRE', 'RCBP AMAPÁ', 
                              'RCBP RONDONIA']),
    df['Nome do RCBP'].isin(['RCBP CAMPINAS-UNICAMP', 'RCBP SÃO PAULO', 'RCBP BELO HORIZONTE', 'RCBP DRS BARRETOS', 
                              'RCBP ESPÍRITO SANTO', 'RCBP SANTOS', 'RCBP ANGRA DOS REIS', 'RCBP POÇOS DE CALDAS', 
                              'RCBP CAMPINAS-SMS', 'RCBP JAHU']),
    df['Nome do RCBP'].isin(['RCBP CURITIBA', 'RCBP PORTO ALEGRE', 'RCBP FLORIANOPOLIS'])
]

choices = ['NORDESTE', 'CENTRO-OESTE', 'NORTE', 'SUDESTE', 'SUL']

df['Regiao'] = np.select(conditions, choices)

# 'RCBP JAHU'
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
output_dir = 'C:/Users/LENOVO/Documents/GitHub/cancer_forecasting/IC-Times-Series/data/parquet/rcbp'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

for value in df['Regiao'].unique():
    sub_df = df[df['Regiao'] == value]
    
    file_name = f'{value}.parquet'
    file_path = os.path.join(output_dir, file_name)
    table = pa.Table.from_pandas(sub_df)
    pq.write_table(table, file_path)
```
