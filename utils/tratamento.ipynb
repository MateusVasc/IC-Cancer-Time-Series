```python
columns_to_drop = ['Código do Paciente', 'Unnamed: 37']

df = df.drop(columns_to_drop, axis=1)
df.columns
```

```python
def atualizar_status(row):
    if pd.notna(row['Tipo do Obito']) or pd.notna(row['Data do Óbito']):
        return 'MORTO'
    else:
        return row['Status Vital']

df['Status Vital'] = df.apply(atualizar_status, axis=1)
```

```python
# df['Data de Nascimento'] = pd.to_datetime(df['Data de Nascimento'], format='%d/%m/%Y')
df['Data do Óbito'] = pd.to_datetime(df['Data do Óbito'], format='%d/%m/%Y')
df['Data de Último Contato'] = pd.to_datetime(df['Data de Último Contato'], format='%d/%m/%Y')
df['Data de Diagnostico'] = pd.to_datetime(df['Data de Diagnostico'], format='%d/%m/%Y')
```

```python
df['Idade'] = pd.to_numeric(df['Idade'], errors='coerce').astype('Int64')
```

```python
conditions = [
    df['Nome do RCBP'].isin(['RCBP SALVADOR', 'RCBP TERESINA', 'RCBP RECIFE', 'RCBP ARACAJU', 'RCBP NATAL', 'RCBP JOAO PESSOA', 
                              'RCBP FORTALEZA', 'RCBP ALAGOAS']),
    df['Nome do RCBP'].isin(['RCBP CAMPO GRANDE', 'RCBP MATO GROSSO (INTERIOR)', 'RCBP CUIABA', 'RCBP DISTRITO FEDERAL', 
                              'RCBP GOIÂNIA']),
    df['Nome do RCBP'].isin(['RCBP RORAIMA', 'RCBP BELÉM', 'RCBP PALMAS', 'RCBP MANAUS', 'RCBP ACRE', 'RCBP AMAPÁ', 
                              'RCBP RONDONIA']),
    df['Nome do RCBP'].isin(['RCBP CAMPINAS-UNICAMP', 'RCBP SÃO PAULO', 'RCBP BELO HORIZONTE', 'RCBP DRS BARRETOS', 
                              'RCBP ESPÍRITO SANTO', 'RCBP SANTOS', 'RCBP ANGRA DOS REIS', 'RCBP POÇOS DE CALDAS', 
                              'RCBP CAMPINAS-SMS', 'RCBP JAHU']),
    df['Nome do RCBP'].isin(['RCBP CURITIBA', 'RCBP PORTO ALEGRE', 'RCBP FLORIANOPOLIS'])
]

choices = ['NORDESTE', 'CENTRO-OESTE', 'NORTE', 'SUDESTE', 'SUL']

df['Regiao'] = np.select(conditions, choices)
```

```python
output_dir = '../data/parquet/'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

df.to_parquet(os.path.join(output_dir, 'dataSetFinalTratado.parquet'))
```

# Tratamento categoria de doenças principais

```python
path3 = '../data/parquet/dataSetFinalTratado.parquet'

df3 = pd.read_parquet(path3)
```

```python
list_mama = ['MAMA , SOE', 'QUADRANTE SUPERIOR EXTERNO DA MAMA', 'PROLONGAMENTO AXILAR DA MAMA', 'QUADRANTE SUPERIOR INTERNO DA MAMA', 'CARCINOMA "IN SITU" DA MAMA, SOE', 'PORCAO CENTRAL DA MAMA', 'QUADRANTE INFERIOR EXTERNO DA MAMA', 'LESAO INVASIVA DA MAMA', 'QUADRANTE INFERIOR INTERNO DA MAMA', 'MAMA']
list_prostata = 'PROSTATA'
list_colo_utero = ['COLO DO UTERO, SOE', 'OUTRAS PARTES DO COLO DO UTERO', 'LESAO INVASIVA DO COLO DO UTERO', 'COLO DO UTERO', 'CARCINOMA "IN SITU" DO COLO DO UTERO (CERVIX)']
list_pele = ['PELE DE OUTRAS PARTES DA FACE, SOE', 'PELE DE OUTRAS PARTES DA FACE', 'PELE DO MEMBRO INFERIOR E DO QUADRIL','PELE DA PALPEBRA E CANTO', 'PELE DO TRONCO','NEOPLASIA MALIGNA DA PELE','PELE DO COURO CABELUDO E DO PESCOCO', 'PELE DA ORELHA E CONDUTO AUDITIVO EXTERNO','PELE DO MEMBRO SUPERIOR E DO OMBRO', 'PELE DOS MEMBROS SUPERIORES, INCLUINDO OMBRO','PELE DO LABIO', 'PELE DA ORELHA E DO CONDUTO AUDITIVO EXTERNO','SARCOMA DE KAPOSI DA PELE', 'PELE DA PALPEBRA, INCLUINDO CANTO','PELE DOS MEMBROS INFERIORES, INCLUINDO QUADRIL', 'MELANOMA MALIGNO DA PELE, SOE','PELE, SOE','LESAO INVASIVA DA PELE', 'MELANOMA MALIGNO INVASIVO DA PELE','PELE DE OUTRAS LOCALIZACOES', 'OUTRAS NEOPLASIAS MALIGNAS DA PELE']
list_pulmao = ['BRONQUIOS OU PULMOES, SOE', 'BRONQUIOS E PULMOES', 'LESAO INVASIVA DOS BRONQUIOS E DOS PULMOES']
```

```python
def categorize_disease(disease, list_mama, list_prostata, list_colo_utero, list_pele, list_pulmao):
    if disease in list_mama:
        return 'Câncer de Mama'
    elif disease in list_prostata:
        return 'Câncer de Próstata'
    elif disease in list_colo_utero:
        return 'Câncer de Colo do Útero'
    elif disease in list_pele:
        return 'Câncer de Pele'
    elif disease in list_pulmao:
        return 'Câncer de Pulmão'
    else:
        return 'Outros'
```

```python
df3['categoria_doenca'] = df3['Descrição da Doenca'].apply(lambda x: categorize_disease(x, list_mama, list_prostata, list_colo_utero, list_pele, list_pulmao))
```

```python
output_dir = '../data/parquet/'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

df3.to_parquet(os.path.join(output_dir, 'dataSetFinalTratadoTop5Diseases.parquet'))
```
